Create a function that calculates the area of a circle and returns it.

`(Area = 3.14 × radius × radius)`

**Output:**
```
Circle area: 78.5
Larger circle area: 314.0
Combined area: 78.26

````python
# Solution:
def circle_area(radius):
    return 3.14 * radius * radius

# Calculate areas
area1 = circle_area(5)
area2 = circle_area(10)
combined_area = area1 + area2

# Print results
print(f"Circle area: {area1}")
print(f"Larger circle area: {area2}")
print(f"Combined area: {combined_area}")
````

**Task: Create a function that converts Celsius to Fahrenheit and returns the result.**

`(Formula: fahrenheit = celsius × 9/5 + 32)`

**Output:**
```
25°C = 77.0°F

````python
# Solution:
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32

# Function calls
temp1 = celsius_to_fahrenheit(25)
freeze = celsius_to_fahrenheit(0)
boil = celsius_to_fahrenheit(100)

# Print results
print(f"25°C = {temp1}°F")
print(f"Water freezes at {freeze}°F")
print(f"Water boils at {boil}°F")
````

**Palindrome Checker**

**Task: Write a function is_palindrome(text) that returns True if the string is the same when reversed, otherwise False.**

````python
# Solution:
def is_palindrome(text):
    return text == text[::-1]

# Example tests
print(is_palindrome("madam"))
print(is_palindrome("hello"))
````

**Multiple Return Values**

**Task: Create a function circle_measurements(radius) that returns both area and circumference of a circle.**

**Use formulas:**

**Area = π × r²**

**Circumference = 2 × π × r**

````python
# Solution:
import math

def circle_measurements(radius):
    area = math.pi * radius ** 2
    circumference = 2 * math.pi * radius
    return area, circumference

# Example usage
area, circumference = circle_measurements(5)

print(f"Area: {area}")
print(f"Circumference: {circumference}")
````

**Task: Create a function that calculates total and average from three game scores and returns both.**

```

Output:

Total: 270, , Average: 90.0
Your total is 280 and average is 93.33333333333333
```

````python
# Solution:
def calculate_scores(score1, score2, score3):
    total = score1 + score2 + score3
    average = total / 3
    return total, average

# Example calls
total1, avg1 = calculate_scores(80, 90, 100)
print(f"Total: {total1}, Average: {avg1}")

total2, avg2 = calculate_scores(90, 95, 95)
print(f"Your total is {total2} and average is {avg2}")
````

**Task: Create a function that finds both the smallest and largest of three numbers.**
```
Output:

Smallest: 5, Largest: 12
Range: 25 to 100
```

````python
# Solution:
def find_smallest_largest(a, b, c):
    smallest = min(a, b, c)
    largest = max(a, b, c)
    return smallest, largest

# Example calls
small1, large1 = find_smallest_largest(5, 12, 9)
print(f"Smallest: {small1}, Largest: {large1}")

small2, large2 = find_smallest_largest(25, 100, 60)
print(f"Range: {small2} to {large2}")
````

**Task: Create a function that takes a full name and returns first name and last name separately.**

**Output:**
```
Name Amrita Prasad
First name: Amrita, Last name: Prasad
Hello Amrita! Your last name is Prasad.

````python
# Solution:
def split_name(full_name):
    parts = full_name.split()
    first_name = parts[0]
    last_name = parts[-1]
    return first_name, last_name

# Example usage
name = "Durga Venkatesh"
print(f"Name {name}")

first, last = split_name(name)
print(f"First name: {first}, Last name: {last}")
print(f"Hello {first}! Your last name is {last}.")
````

**Task: Create a function that finds the highest number from any amount of scores**

````python
# Solution:
def highest_score(*scores):
    return max(scores)

# Example calls
print(highest_score(65, 78, 92, 85))
print(highest_score(100, 98, 89))
print(highest_score(45, 60))
````

**Task: Create a function that calculates the average of any amount of numbers**

````python
# Solution:
def calculate_average(*numbers):
    return sum(numbers) / len(numbers)

# Example calls
print(calculate_average(10, 20, 30))
print(calculate_average(5, 15, 25, 35))
print(calculate_average(2, 4))
````
